## 1. Getting Started - Variables and Constants

```python
# Import everything from SageMath
from sage.all import *

print("=== AFT Learning Notebook 1 ===")
print("Basic SageMath for Five-Operator Theory")
print("Christopher R. Amon - Amon Research Labs")
print()

# Define basic variables
x, y, z, t = var('x y z t', real=True)
print("Spatial and time variables defined: x, y, z, t")

# Physical constants (we'll use symbolic for now)
hbar, c, m, k = var('hbar c m k', positive=True)
print("Physical constants defined: â„, c, m, k")
print()

print("âœ… SageMath environment ready for AFT work!")
```

## 2. Simple Field Functions
Let's create some basic field patterns that will show up in AFT:

```python
# Simple Gaussian field - basic morphogenic pattern
gaussian_field = exp(-(x^2 + y^2)/2)
print("Gaussian field: Ïˆâ‚ =", gaussian_field)
print()

# Oscillating field - wave-like behavior
wave_field = cos(x) * sin(y) * exp(-t/4)
print("Wave field: Ïˆâ‚‚ =", wave_field)
print()

# Combined field - more complex morphogenic pattern
morphic_field = gaussian_field * cos(sqrt(x^2 + y^2))
print("Morphic field: Ïˆâ‚ƒ =", morphic_field)
print()

print("âœ… Basic field patterns created!")
print("These represent different types of substrate patterns.")
```

## 3. The Five Operators - Symbolic Form
Let's define our five operators symbolically:

```python
print("=== THE FIVE OPERATORS (Symbolic) ===")
print()

# We'll work with a test field
psi = function('psi')(x, y, t)
print("Test field: Ïˆ(x,y,t)")
print()

# Point Operator PÌ‚ - creates localization
print("PÌ‚: Point Operator - Existence/Localization")
print("   PÌ‚Ïˆ = âˆ« Î´(r - r') Ïˆ(r') dr'")
print("   (Localizes field at specific points)")
print()

# Line Operator LÌ‚ - creates connections  
L_op_result = -I*hbar*c*diff(psi, x)  # Simplified 1D version
print("LÌ‚: Line Operator - Connection/Direction")
print("   LÌ‚Ïˆ = -iâ„câˆ‡Ïˆ â‰ˆ", L_op_result, "(1D version)")
print("   (Creates momentum, connections)")
print()

# Curve Operator Äˆ - creates curvature
C_op_result = -hbar^2*c^2/(2*m)*diff(psi, x, 2)
print("Äˆ: Curve Operator - Curvature/Acceleration")
print("   ÄˆÏˆ = -â„Â²cÂ²/2m âˆ‡Â²Ïˆ â‰ˆ", C_op_result, "(kinetic part)")
print("   (Creates bending, focusing)")
print()

# Movement Operator MÌ‚ - creates dynamics
M_op_result = -I*hbar*c*diff(psi, x)
print("MÌ‚: Movement Operator - Dynamics/Evolution")
print("   MÌ‚Ïˆ = -iâ„câˆ‡Ïˆ â‰ˆ", M_op_result)
print("   (Creates motion, evolution)")
print()

# Resistance Operator RÌ‚ - creates stability
R_op_result = m*c^2*psi
print("RÌ‚: Resistance Operator - Constraint/Stability")
print("   RÌ‚Ïˆ = mcÂ²Ïˆ =", R_op_result)
print("   (Provides mass, inertia)")
print()

print("âœ… Five operators defined symbolically!")
```

## 4. Simple Calculations - Derivatives and Operations
Practice basic calculus operations that show up everywhere in AFT:

```python
print("=== BASIC CALCULUS PRACTICE ===")
print()

# Work with our simple Gaussian field
test_field = exp(-(x^2 + y^2)/2)
print("Working with: Ïˆ =", test_field)
print()

# First derivatives (gradient components)
dpsi_dx = diff(test_field, x)
dpsi_dy = diff(test_field, y)
print("âˆ‚Ïˆ/âˆ‚x =", dpsi_dx)
print("âˆ‚Ïˆ/âˆ‚y =", dpsi_dy)
print()

# Second derivatives (Laplacian components)
d2psi_dx2 = diff(test_field, x, 2)
d2psi_dy2 = diff(test_field, y, 2)
laplacian = d2psi_dx2 + d2psi_dy2
print("âˆ‚Â²Ïˆ/âˆ‚xÂ² =", d2psi_dx2)
print("âˆ‚Â²Ïˆ/âˆ‚yÂ² =", d2psi_dy2)
print("âˆ‡Â²Ïˆ = âˆ‚Â²Ïˆ/âˆ‚xÂ² + âˆ‚Â²Ïˆ/âˆ‚yÂ² =", laplacian)
print()

# Simplify the Laplacian result
laplacian_simplified = simplify(laplacian)
print("âˆ‡Â²Ïˆ simplified =", laplacian_simplified)
print()

print("âœ… Basic derivatives calculated!")
print("These are the building blocks of all AFT operators.")
```
